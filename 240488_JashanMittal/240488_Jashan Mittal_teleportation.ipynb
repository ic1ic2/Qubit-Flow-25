```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import DensityMatrix, partial_trace
from qiskit.visualization import plot_state_city
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
qc=QuantumCircuit(3)
qc.rx(1,0) #to randomize q0
#Just to get the initial state of Alice
qc1=qc.copy()
print("density matrix of alice")
rho1=DensityMatrix.from_instruction(qc1)
rho1=partial_trace(rho1, [1,2])
dm1=rho1.data
print(dm1)
#circuit
qc.h(1)
qc.cx(1,2)
qc.cx(0,1)
qc.h(0)
qc.cx(1, 2)
qc.cz(0, 2)
print("CIRCUIT DIAGRAM")
qc.draw('mpl')
```

```python

rho_full = DensityMatrix.from_instruction(qc)
rho_q2=partial_trace(rho_full, [0,1])
dm=rho_q2.data
print("Final state of Bob")
print(dm)
```

**Density matrix of Alice initially is same as the matrix of Bob's final state signalling complete teleportation**
