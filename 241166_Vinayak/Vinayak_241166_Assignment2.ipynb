```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import numpy as np
from qiskit.visualization import plot_histogram
```

2-spin

```python
J = 1.0
time = np.pi

qc2 = QuantumCircuit(2)
qc2.h([0,1])  

qc2.rxx(2*time*J, 0, 1)
qc2.ryy(2*time*J, 0, 1)
qc2.rzz(2*time*J, 0, 1)

qc2.measure_all()  

qc2.draw('mpl')

sim2 = AerSimulator()
result2 = sim2.run(qc2,shots=1000).result()
```

```python
print(result2.get_counts())

plot_histogram(result2.get_counts())
```

3-spin

```python
J = 1.0
time = np.pi

qc3 = QuantumCircuit(3)
qc3.h([0,1,2]) 
qc3.rxx(2*time*J, 0, 1)
qc3.ryy(2*time*J, 0, 1)
qc3.rzz(2*time*J, 0, 1)

qc3.rxx(2*time*J, 1, 2)
qc3.ryy(2*time*J, 1, 2)
qc3.rzz(2*time*J, 1, 2)

qc3.rxx(2*time*J, 0, 2)
qc3.ryy(2*time*J, 0, 2)
qc3.rzz(2*time*J, 0, 2)

qc3.measure_all()  
```

```python

sim3 = AerSimulator()
result3 = sim3.run(qc3, shots=1000).result()
print(result3.get_counts())

plot_histogram(result3.get_counts())
```
