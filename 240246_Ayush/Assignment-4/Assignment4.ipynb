```python
#Applying single trotter step
def heisenberg_trotter_step(qc, delta_t, J=1.0):
    qc.rxx(2 * J * delta_t, 0, 1)  # XX term
    qc.ryy(2 * J * delta_t, 0, 1)  # YY term
    qc.rzz(2 * J * delta_t, 0, 1)  # ZZ term
    return qc

#just generates the circuit at time t
def timeEvolution(T, n_steps):
    qc = initial_state()
    delta_t = T / n_steps
    for _ in range(n_steps):
        heisenberg_trotter_step(qc, delta_t)
    return qc

#Actual function to find fidelity
def compute_fidelity(ideal_qc, noisy_counts):
    ideal_state = Statevector(ideal_qc)
    total_shots = sum(noisy_counts.values())
    return sum(
        np.abs(ideal_state[int(state, 2)])**2 * (count/total_shots)
        for state, count in noisy_counts.items()
    )

def run_noisy_simulation(qc, shots=1000):
    measured_qc = qc.copy()
    measured_qc.measure_all()
    t_qc = transpile(measured_qc, fake_backend)
    result = noisy_sim.run(t_qc, shots=shots).result()
    return result.get_counts()


times = np.linspace(0, 2*np.pi, 20)  # Time points
n_steps = 10                         # Trotter steps
shots = 1024                         # Shots per simulation
fidelities = []
for t in times:
    ideal_circuit = timeEvolution(t, n_steps)
    noisy_counts = run_noisy_simulation(ideal_circuit, shots)
    fid = compute_fidelity(ideal_circuit, noisy_counts)
    fidelities.append(fid)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(times, fidelities, 'ro-', )
plt.xlabel('Time', fontsize=12)
plt.ylabel('Fidelity', fontsize=12)
plt.title('Fidelity Score')
plt.axhline(1.0, color='k', ls='--', alpha=0.5)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
