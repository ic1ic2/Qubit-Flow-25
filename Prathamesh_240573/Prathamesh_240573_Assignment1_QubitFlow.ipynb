```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
import numpy as fufu

dudu = QuantumRegister(3, 'dudu')
zuzu = ClassicalRegister(2, 'zuzu')
mumu = QuantumCircuit(dudu, zuzu)

theta, phi = fufu.pi/4, fufu.pi/2
mumu.initialize([fufu.cos(theta/2), fufu.exp(1j*phi)*fufu.sin(theta/2)], dudu[0])

mumu.h(dudu[1])
mumu.cx(dudu[1], dudu[2])

mumu.cx(dudu[0], dudu[1])
mumu.h(dudu[0])

mumu.measure(dudu[0], zuzu[0])
mumu.measure(dudu[1], zuzu[1])

with mumu.if_test((zuzu[0], 1)):
    mumu.z(dudu[2])
with mumu.if_test((zuzu[1], 1)):
    mumu.x(dudu[2])

print("Quantum Teleportation Circuit:")
mumu.draw(output='mpl', style='iqp')
```
